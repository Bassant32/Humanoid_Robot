Model Training Setup & Hyperparameters

This notebook documents the configuration and hyperparameters used to train the YOLO object detection model. The goal was to achieve stable convergence, good generalization, and efficient training using transfer learning and data augmentation.

```python
output_path = "/content/drive/MyDrive/Dataset/results/"
import os
os.makedirs(output_path, exist_ok=True)
```

```python
from ultralytics import YOLO

model = YOLO("yolo11n.pt")   # Nano
```

```python
model.train(
    data="/content/drive/MyDrive/Dataset/data.yaml",
    epochs=80,
    imgsz=640,
    batch=16,
    lr0=0.001,
    optimizer="SGD",
    patience=20,
    freeze=10,
    weight_decay = 0.0005,
    workers=2,
    device=0,
     project="/content/drive/MyDrive/Dataset/results",
    name="training_run",
    exist_ok=True,


    augment=True,
    mosaic=1.0,
    mixup=0.1,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    translate=0.1,
    scale=0.5,
    fliplr=0.5
)
```

 **Training Hyperparameters Summary**

Epochs: 80 â€” sufficient number of iterations to ensure convergence without overtraining.

Image size (imgsz): 640 â€” balances detection accuracy and computational cost.

Batch size: 16 â€” chosen to fit GPU memory while maintaining stable gradient updates.

Optimizer: SGD â€” selected for its stability and better generalization compared to adaptive optimizers.

Initial learning rate (lr0): 0.001 â€” low learning rate to avoid unstable training.

Weight decay: 0.0005 â€” regularization term to reduce overfitting.

Early stopping patience: 20 â€” stops training when validation performance stops improving.

Frozen layers: 10 â€” enables transfer learning by preserving pretrained backbone features.

Workers: 2 â€” number of parallel data loading processes.

Device: GPU (device 0) â€” used to accelerate training.

** Augmentation Hyperparameters **

Augment: Enabled â€” applies data augmentation during training.

Mosaic: 1.0 â€” combines multiple images to improve small object detection.

MixUp: 0.1 â€” blends images to enhance robustness.

HSV Hue: 0.015 â€” slight color variation for lighting robustness.

HSV Saturation: 0.7 â€” increases color diversity.

HSV Value: 0.4 â€” adjusts brightness variations.

Translate: 0.1 â€” random spatial shifting.

Scale: 0.5 â€” random resizing for scale invariance.

Flip (horizontal): 0.5 â€” horizontal flipping probability.

** Modified Hyperparameters (Stability Improvement)**

Initial learning rate (lr0): increased to 0.01 due to high oscillations and unstable loss behavior.

Mosaic augmentation: reduced to 0.5 to decrease excessive image distortion and training noise.
